## Вычисление pi с помощью метода Монте-Карло

# Формулы:
-площадь круга $s = (2r)^2$
-площадь квадрата $c = \pi r^2$
-$c/s = (\pi r^2)/(4 r^2) = \pi / 4$
-$\pi = 4 * c/s$

# Код

```python
import random
from ipywidgets import interact, widgets

%matplotlib inline
import matplotlib.pyplot as plt

def throw_darts(num_points):
    points = []
    hits = 0
    for _ in range(num_points):
        x, y = random.random(), random.random()
        if x*x +  y*y < 1.0:
            hits += 1
            points.append((x,y, "red"))
        else:
            points.append((x,y, "blue"))
    
    fraction = hits / num_points
    pi = 4*fraction
    return pi, points

def create_plot(points):
    x, y, colors = zip(*points)
    plt.scatter(x, y, c = colors)

def experiment(num_points):
    pi, points = throw_darts(num_points)
    create_plot(points)
    print("approximation", pi)
```

```python
interact(experiment, num_points = widgets.IntSlider(min = 100, max = 10000, step = 100, value = 4900))
```
